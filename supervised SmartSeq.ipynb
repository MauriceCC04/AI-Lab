```python
#function to add new 'Oxia' row to the data frame
#1 is Normoxia, 0 is Hypoxia
def oxia(df):
    new_row = {col: None for col in df.columns}
    for col in df.columns:
        if '_Nor' in col:
            new_row[col] = 1
        elif '_Hyp' in col:
            new_row[col] = 0
    df = pd.concat([df, pd.DataFrame(new_row, index=['oxia'])])
    return df
```

```python
#it worked now for the y
y_mcf = MCF.loc['oxia']
y_hcc = HCC.loc['oxia']
```

```python
#do the same model selection as in drop seq, first with MCF then HCC
#first split the data
X_train_MCF, X_test_MCF, y_train_MCF, y_test_MCF = train_test_split(X_MCF.T, y_mcf, test_size=0.2, random_state=42)
X_train_HCC, X_test_HCC, y_train_HCC, y_test_HCC = train_test_split(X_HCC.T, y_hcc, test_size=0.2, random_state=42)
```

```python
#fit the models, first linear regression, then logistic regression, then KNN, then random forest then SVM
#MCF
#Linear Regression
lin_reg_MCF = LinearRegression()
lin_reg_MCF.fit(X_train_MCF, y_train_MCF)
y_pred_MCF = lin_reg_MCF.predict(X_test_MCF)
y_pred_MCF = np.round(y_pred_MCF)
```

```python
#accuracy, mean squared error, confusion matrix
print("Accuracy: ", accuracy_score(y_test_MCF, y_pred_MCF))
print("Confusion Matrix: ", confusion_matrix(y_test_MCF, y_pred_MCF))
print("Mean Squared Error: ", mean_squared_error(y_test_MCF, y_pred_MCF))
print("R2 Score: ", r2_score(y_test_MCF, y_pred_MCF))
```

```python
#Logistic Regression
log_reg_MCF = LogisticRegression(solver='liblinear')
params = {'penalty': ['l1', 'l2'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
log_reg_MCF.fit(X_train_MCF, y_train_MCF)
#using gridsearch to find the best parameters
log_reg_gs_MCF = GridSearchCV(log_reg_MCF, params, cv=5, scoring=['neg_log_loss'], refit='neg_log_loss')
log_reg_gs_MCF.fit(X_train_MCF, y_train_MCF)
```

```python
#best parameters
print(f"The best parameters for the logistic regression:{log_reg_gs_MCF.best_params_}")
```

```python
#predictions with the best parameters
# Get the best model
best_log_MCF = log_reg_gs_MCF.best_estimator_

# Use the best model to make predictions
y_pred_MCF = best_log_MCF.predict(X_test_MCF)

# Calculate accuracy
accuracy_logistic_MCF = accuracy_score(y_test_MCF, y_pred_MCF)
print(f"Accuracy: {accuracy_logistic_MCF}")
print(f"Confusion Matrix: {confusion_matrix(y_test_MCF, y_pred_MCF)}")
print(f"Mean Squared Error: {mean_squared_error(y_test_MCF, y_pred_MCF)}")
print(f"R2 Score: {r2_score(y_test_MCF, y_pred_MCF)}")
```

```python
#k nearest neighbours
knn_MCF = KNeighborsClassifier()
params = {'n_neighbors': [3, 5, 7, 9, 11], 'weights': ['uniform', 'distance']}
knn_MCF.fit(X_train_MCF, y_train_MCF)
knn_gs_MCF = GridSearchCV(knn_MCF, params, cv=5, scoring=['neg_log_loss'], refit='neg_log_loss')
knn_gs_MCF.fit(X_train_MCF, y_train_MCF)
best_knn_MCF = knn_gs_MCF.best_estimator_
```

```python
#best parameters
best_knn_MCF.fit(X_train_MCF, y_train_MCF)
y_pred_MCF = best_knn_MCF.predict(X_test_MCF)
accuracy_knn = accuracy_score(y_test_MCF, y_pred_MCF)
print(f"Accuracy: {accuracy_knn}")
print(f"Confusion Matrix: {confusion_matrix(y_test_MCF, y_pred_MCF)}")
print(f"Mean Squared Error: {mean_squared_error(y_test_MCF, y_pred_MCF)}")
print(f"R2 Score: {r2_score(y_test_MCF, y_pred_MCF)}")
```

```python
#random forest
forest_MCF = RandomForestClassifier(random_state=42)
params = {'n_estimators': [10, 50, 100, 200], 'max_depth': [10, 50, 100, 200]}
forest_MCF.fit(X_train_MCF, y_train_MCF)
forest_gs_MCF = GridSearchCV(forest_MCF, params, cv=5, scoring=['neg_log_loss'], refit='neg_log_loss')
forest_gs_MCF.fit(X_train_MCF, y_train_MCF)
best_forest_MCF = forest_gs_MCF.best_estimator_
best_forest_MCF.fit(X_train_MCF, y_train_MCF)
```

```python
y_pred_MCF = best_forest_MCF.predict(X_test_MCF)
accuracy_forest_MCF = accuracy_score(y_test_MCF, y_pred_MCF)
print(f"Accuracy: {accuracy_forest_MCF}")
print(f"Confusion Matrix: {confusion_matrix(y_test_MCF, y_pred_MCF)}")
print(f"Mean Squared Error: {mean_squared_error(y_test_MCF, y_pred_MCF)}")
print(f"R2 Score: {r2_score(y_test_MCF, y_pred_MCF)}")
```

```python
#Support Vector Machine
svm_MCF = SVC(probability=True)
params = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000], 'kernel': ['linear', 'poly', 'rbf', 'sigmoid']}
svm_MCF.fit(X_train_MCF, y_train_MCF)
svm_gs_MCF = GridSearchCV(svm_MCF, params, cv=5, scoring=['neg_log_loss'], refit='neg_log_loss')
svm_gs_MCF.fit(X_train_MCF, y_train_MCF)
best_svm_MCF = svm_gs_MCF.best_estimator_
best_svm_MCF.fit(X_train_MCF, y_train_MCF)
```

```python
y_pred_MCF = best_svm_MCF.predict(X_test_MCF)
accuracy_svm_MCF = accuracy_score(y_test_MCF, y_pred_MCF)
print(f"Accuracy: {accuracy_svm_MCF}")
print(f"Confusion Matrix: {confusion_matrix(y_test_MCF, y_pred_MCF)}")
print(f"Mean Squared Error: {mean_squared_error(y_test_MCF, y_pred_MCF)}")
print(f"R2 Score: {r2_score(y_test_MCF, y_pred_MCF)}")
```
